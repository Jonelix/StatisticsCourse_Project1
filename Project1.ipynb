```python
import numpy as np
```

# TOOLS

```python
# N_DICE
def n_dice(n: int) -> list[float]:
    return (np.ones(n) / n).tolist()
```

# TASK 1 & 2

```python
# Create the dice

dice4 = n_dice(4)
dice6 = n_dice(6)
dice8 = n_dice(8)
dice12 = n_dice(12)
dice20 = n_dice(20)

# Convolve the dice

convolution = np.convolve(dice4, dice6)
convolution = np.convolve(convolution, dice8)
convolution = np.convolve(convolution, dice12)
convolution = np.convolve(convolution, dice20)

print("Length of convolution:", len(convolution))
print(convolution)
```

```python
# Sum indexes 0 to 5
sum_0_5 = sum(convolution[0:6])

# Sum indexes 41 to 46
sum_41_46 = sum(convolution[41:47])

# Sum of all indexes
sum_all = sum(convolution)

probability_win = (sum_0_5 + sum_41_46) / sum_all

print("Probability of winning:", probability_win)
```

```python
import matplotlib.pyplot as plt

# Plot
plt.bar(range(len(convolution)), convolution, width=0.6, edgecolor="black")
plt.title(f"Probability Distribution of a N-sided Die")
plt.xlabel("Die Face")
plt.ylabel("Probability")
plt.show()
```

```python
from numpy import random

NumberOfTrials = 1000
RecordResult = 0

for i in range(NumberOfTrials):
    # Simulate the game
    if random.random() < probability_win:
        RecordResult += 1
```
